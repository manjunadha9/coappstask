```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pokemon_data.csv")
df.head(10)
```

```python
df.isnull().sum()
```

```python
corr = df.corr()
corr
```

```python
plt.imshow(corr, cmap="rainbow")

plt.colorbar()

variables = []
for i in corr.columns:
    variables.append(i)
    
plt.xticks(range(len(corr)), variables, rotation = 45)
plt.yticks(range(len(corr)), variables)

plt.show()
```
